```python
import fastf1
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Enable FastF1 cache
fastf1.Cache.enable_cache('f1_cache')

# Load the session
session = fastf1.get_session(2018, 'Azerbaijan', 'Q')
session.load()

# Get weather data
weather_data = session.weather_data

# Display basic info about the weather data
print("Weather Data Columns:", weather_data.columns)
print("\nFirst few rows of weather data:")
print(weather_data.head())
```
